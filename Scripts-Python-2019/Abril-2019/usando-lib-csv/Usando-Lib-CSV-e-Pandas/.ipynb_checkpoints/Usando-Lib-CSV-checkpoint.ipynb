```python
''''
Tema: Deletar um intervalo de linha e coluna no python?
Problema no Stackoverflow: https://repl.it/@moisestedeschi/Deletar-um-intervalo-de-linha-e-coluna-no-python
@author: Moisés Tedeschi
Created: 17/04/2019
'''

import csv

plan_in = open('nome-sobrenome.csv', 'r')
with open('nome-sobrenome - novo.csv', 'w', newline="") as csvfile:
    
    data_in_csv = csv.reader(plan_in)
    data_out_csv = csv.writer(csvfile)
    
    for row_number, row in enumerate(data_in_csv):
        #SE nr linha for igual a 0 (cabeçalho) ou nr linha estiver entre o intervado desejado:
        
        if row_number == 0 or ( row_number  >= 1 and row_number  <= 60 ): 
            data_out_csv.writerow(row[:])
            #print('Row #' + str(data_in_csv.line_num) + '' + str(str(row)) + '\n')
        
plan_in.close()
csvfile.close()
```
