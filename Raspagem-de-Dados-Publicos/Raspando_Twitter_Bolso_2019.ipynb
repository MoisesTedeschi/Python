```
from twitter_scraper import get_tweets

cont = 0
print ('Raspando os últimos 500 Tweets de uma conta em segundos. \n')
conta_twitter = input('Insira o nome da conta: ')
for tweets in get_tweets(conta_twitter): 
    print ('\nID do Tweet: ', tweets['tweetId'])
    print ('Mensagem: ',tweets['text'])
    print ('Toral de curtidas: ', tweets['likes'])
    print ('Total de retweets: ', tweets['retweets'])
    print ('Total de Respostas: ', tweets['replies'])
    print ('Conteúdo complementar: ', tweets['entries'])
    print ('Link do Tweet: https://twitter.com/'+conta_twitter+'/status/'+tweets['tweetId'])
    print ('\n**********\n')
    cont += 1

print(f'Total de Tweets raspados: {cont}')
```
