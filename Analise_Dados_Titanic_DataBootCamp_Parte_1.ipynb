Na coluna `Cabin`, o valor (NaN) se repete várias vezes ao longo das linhas. Assim sendo, vamos analisar o fato com mais calma logo abaixo. Valores "NaN" influenciam no cálculo dos quartis e mediana (idem as linhas "25%", "50%" e "75%").

##1.1 - Pergunta: Qual a porcentagem de sobreviventes e NÃO sobreviventes?

```python
# Porcentagem de sobreviventes com base no total de tripulantes.
def calcular_porcentagem(val, total):
    """Calcula porcentagem Titanic"""

    porcentagem = float(val / total)
    result_porcentagem = ("%.2f" % (porcentagem * 100) + "%")
    return result_porcentagem
```

```python
# Pegando o total de tripulantes do Titanic.
total_tripulantes = titanic.Survived.count()

nao_sobreviveu = titanic.Survived[titanic["Survived"] == 0].count()
sobreviveu = titanic.Survived[titanic["Survived"] == 1].count()
```

```python
print("Tripulantes que sobreviveram: ",calcular_porcentagem(sobreviveu, total_tripulantes))
print("Tripulantes que NÃO sobreviveram: ",calcular_porcentagem(nao_sobreviveu, total_tripulantes))
```

```python
print('Sobreviventes     :', titanic['Survived'].value_counts()[1])
print('Não Sobreviventes :', titanic['Survived'].value_counts()[0])
```

```python
colors_pie = ['lightcoral', 'lightskyblue']
titanic['Survived'].value_counts().plot.pie( colors=colors_pie,
  title='Qual a porcentagem de sobreviventes e NÃO sobreviventes?', fontsize=18,
  shadow=True, startangle=90, autopct='%1.2f%%',
  labels=('Não sobreviventes','Sobreviventes' ), figsize=(10,10)).set_ylabel('');
```

### Conlusão:
Apenas 38,38% dos passageiros do Titanic sobreviveram a tragédia. Enquanto que quase o dobro dos passageiros não conseguiu sobreviver (61,62%).

##1.2 - Pergunta: Como estão divididos esses passageiros sobreviventes e NÃO sobreviventes?

```python
# Quantidade de sobreviventes e mostros no naufrágio. 
plt.figure(figsize=(16,10))
tripulantes = sns.countplot(data=titanic, x='Survived', hue='Sex', palette='Reds_d')

# Colocando valores acima das barras
# Pode-se acessar a altura da barra e informações sobre elas "patches".
for tripulante in tripulantes.patches:
    
    # Acessado os objetos de "patch" associados ao text.
    barras = tripulante.get_bbox()
    # Adicionando os valores dos
    tripulantes.annotate("{:.0f}".format(barras.y1 + barras.y0), ((barras.x0 + barras.x1) / 2 - 0.03, barras.y1 + 15))

plt.title('QUANTIDADE DE SOBREVIVENTES E NÃO SOBREVIVENTES\n TOTAL: %s TRIPULANTES' % (titanic.shape[0]))
plt.xlabel('SOBREVIVEU OU NÃO?')
plt.ylabel('NÚMERO DE SOBREVIVENTES')
plt.show()
```

## Mas pq isso?

No meu entendimento `SibSp`(Quantidade de cônjuges e irmãos a bordo) e `Parch`(Quantidade de pais e filhos a bordo) são colunas que possuem relação direta com o passageiro. Os homens têm uma alta probabilidade de sobrevivência se estiverem na porta `"C"`, mas uma baixa probabilidade se estiverem na porta `"Q"` ou `"S"`. Como vimos antes, o `Pclass` tem correlação com a sobrevivência.

## Birkenhead Drill

<a href="https://pt.wikipedia.org/wiki/Mulheres_e_crianças_primeiro">"Mulheres e crianças primeiro"</a> (ou em menor grau, o ***Birkenhead Drill***) é um código de conduta datando de 1852, pelo qual as vidas de mulheres e crianças deveriam ser salvas primeiro em uma situação que ameaça a vida, geralmente abandonando o navio, quando os recursos de sobrevivência, como botes salva-vidas são limitados. Sabemos que o acidente do Titanic ocorreu em 1912, décadas depois da criação deste código, que protocolava a ordem de resgate de uma situação de vida e morte. 
<br />
Enquanto a frase aparece pela primeira vez no romance de 1860 Harrington: A Story of True Love, de William Douglas O'Connor, a primeira aplicação documentada de "mulheres e crianças primeiro" ocorreu em 1852 durante a evacuação do navio da Marinha Real Britânica HMS Birkenhead. No entanto, é mais associado ao naufrágio do RMS Titanic em 15 de abril de 1912. Como código de conduta, "mulheres e crianças primeiro" não tem base no direito marítimo. De acordo com o especialista em evacuação de desastres Ed Galea, em evacuações nos tempos atuais as pessoas geralmente ajudarão os mais vulneráveis a abandonar a cena primeiro, provavelmente os feridos, idosos e crianças pequenas."

##VI. Conclusão

O trágico naufrágio do navio transatlântico *RMS Titanic*, ainda hoje, é considerado uma das maiores tragédias náuticas da história. Muitos relatos históricos do naufrágio do Titanic descrevem que o navio de passageiros com quase 270 metros *“deslizou por baixo das ondas do mar”*, como se a embarcação e seus passageiros tivessem sido arrastados tranquilamente para o repouso, o que não poderia estar mais longe da verdade.

O destino do Titanic foi selado em sua viagem inaugural de Southampton, na Inglaterra, à cidade de Nova York. Às 23h40 de 14 de abril de 1912, a lateral do Titanic colidiu com um iceberg no norte do Atlântico, afundando partes do casco do estibordo por uma extensão de quase 100 metros e expondo à água do mar os seis compartimentos dianteiros à prova d’água. A partir deste fato, o naufrágio foi inevitável.

Embora o navio transatlântico tivesse capacidade para transportar 3.511 passageiros, o Titanic levava apenas botes salva-vidas suficientes para 1.178 pessoas. Para piorar ainda mais, alguns botes salva-vidas não foram preenchidos à capacidade máxima durante a evacuação desesperada dos passageiros do RMS Titanic. A maioria das cerca de 1,5 mil vítimas do Titanic morreu de hipotermia na superfície da água gelada. Centenas ainda podem ter morrido dentro do navio quando afundou, a maioria formada por famílias de imigrantes da 3º classe (Pobres), que ansiavam por “oportunidades” e por uma vida melhor nos Estados Unidos.

Além das vidas perdidas no naufrágio, algo mais se perdeu com o Titanic. O "algo mais" pode ser relacionando diretamente a ilusão de organização, a fé no progresso tecnológico, um anseio pelo futuro que, quando a Europa foi arrastada pelas circunstâncias a uma ampla guerra, logo foi substituído pelos medos e temores tão familiares do nosso mundo moderno.

Segundo o cineasta James Cameron, “O desastre do Titanic foi como a explosão de uma bolha. Havia uma forte sensação de abundância na primeira década do século 20. Elevadores! Automóveis! Aviões! Rádio sem fio! Tudo parecia tão maravilhoso, uma evolução infinita. Então tudo veio abaixo”.

Diante da análise dos dados, os principais resultados que visualizamos foi a grande quantidade de pessoas que não sobreviveram ao desastre. Embora a “sorte” pareça ter sorrido para aqueles mais abastados - O fato da 1º classe (Ricos) ter tido uma proporção maior de sobreviventes e acesso mais rápido aos botes salva-vidas, não quer dizer que as perdas não foram substanciais nas três classes da tripulação do Titanic. As mortes foram sentidas indiferente do gênero (mesmo com o fator "Birkenhead Drill"), idade, classe social, cor, nacionalidade e/ou status. Com isso, até hoje, o trágico naufrágio do Titanic é motivo de estudos para várias áreas das ciências humanas, exatas e comportamentais.
